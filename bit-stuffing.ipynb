```python
def bit_stuffing(data,flag,bitpattern):
    stuffed_data = ''
    count = 0
    
    for i in range(len(data)):
        
        stuffed_data += data[i]
        
        if data[i:i+len(bitpattern)] == bitpattern:
            stuffed_data += '0'
            count += 1
            
    stuffed_data = flag + stuffed_data + flag
    
    return stuffed_data,count

data = input('enter data -> ')
flag = input('enter flag ->')
bitpattern = input('enter bitpattern ->')

stuffed_data,count = bit_stuffing(data,flag,bitpattern)

print("Stuffed data:", stuffed_data)
print("Stuffed bit count:", count)
```
